```python
stars=pd.read_csv('/kaggle/input/stars-csv/Stars.csv')
stars.info()
stars.describe().columns
```

```python
def plot_star_temperatures(csv_file=stars):
    data = pd.read_csv('/kaggle/input/stars-csv/Stars.csv')

    # Check if necessary columns exist
    if 'Star type' not in data.columns or 'Temperature (K)' not in data.columns:
        print("Error: CSV must contain 'Star type' and 'Temperature' columns.")
        return

    # Sort the data if needed (optional)
    data = data.sort_values(by='Star type')

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.bar(data['Star type'], data['Temperature (K)'], color='orange', linewidth=0.5)

    # Labels and Title
    plt.xlabel('Star Type')
    plt.ylabel('Temperature (K)')
    plt.title('Temperature of Stars by Star Type')
    plt.grid(True)

    # Show the plot
    plt.show()
plot_star_temperatures()
```
